```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#load data(sometimes need to use "/")
dfpos = pd.read_csv('D:/pythonData/Confirm/dementia/Positive_Likelihood.csv')
dfneg = pd.read_csv('D:/pythonData/Confirm/dementia/Negative_Likelihood.csv')
```

```python
#plot bar chart
sns.set(style='darkgrid')
plt.figure(figsize=(20,10),dpi=600)
plt.axvline(x=0.32,color='black',linestyle='--',linewidth=1,label='Threshold:0.32')
sns.kdeplot(x='positive', data=dfpos,shade=True,cut=0,color='red',label='Positive Likelihood distribution')
sns.kdeplot(x='positive', data=dfneg,shade=True,cut=0,color='blue',label='Negative Likelihood distribution')

plt.xlim(-0.05,1.05)
# axis labels
plt.xlabel('Predicted Likelihood',fontsize=18)
plt.ylabel('Density',fontsize=18)
plt.legend(fontsize=16)
plt.show()
```

```python
#plot bar chart
sns.set(style='darkgrid')
plt.figure(figsize=(20,10),dpi=600)
sns.histplot(x='positive', data=dfpos,color='red',kde=True,stat='probability', label='Positive likelihood')
sns.histplot(x='positive', data=dfneg,color='blue',kde=True, stat='probability', label='Negative likelihood')

# axis labels
plt.xlabel('Predicted Likelihood',fontsize=18)
plt.ylabel('Probability',fontsize=18)
plt.legend(fontsize=16)
plt.show()
```
